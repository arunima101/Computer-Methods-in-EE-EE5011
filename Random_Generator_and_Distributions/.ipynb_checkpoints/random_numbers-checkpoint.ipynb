```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
from scipy.special import chdtrc
```

```python
def generate_normal_deviate(i):   ### Implementing Box-muller algo in Book.
    y = []
    for _ in range(i):
        rsq,v1,v2 =0,0,0    
        while (rsq>=1 or rsq ==0):
            x1,x2 = np.random.uniform(0,1,2)
            v1 = 2*x1 -1
            v2 = 2*x2 -1
            rsq = v1**2 + v2**2
        fac = np.sqrt(-2*np.log(rsq)/rsq)
        y.append(v2*fac)
    return y


y= generate_normal_deviate(50000)



fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(y,bins=100,color='green')
plt.show()
```

```python
##### Question 2 #####

changing_deviates = [20,50,100,200,300,400,500,600,700,800,900,1000]
bin_size = 1
bins = np.arange(-3,4,bin_size)
no_of_deviates = changing_deviates[np.random.randint(0,len(changing_deviates))]
# bins_ = np.arange(-3,4,1)

y= generate_normal_deviate(no_of_deviates)

    

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(y,bins=bins,color='green')
plt.title(f"{no_of_deviates} deviates bin size {bin_size}")
plt.grid(True)
plt.show()
```

```python
###### Chi-square test #####
def chi_square_test(bins,y):
    f_obs =[0]*len(bins)
    for j in range(len(bins)-1):
        for i in y:
            if (i>=bins[j] and i<=bins[j+1]):
                f_obs[j]+=1

    f_exp = np.mean(f_obs)
    
    chi_val = 0
    for i in range(len(f_obs)):
        temp = (f_obs[i]-f_exp)**2/f_exp

        chi_val +=temp
        
    return chi_val, chdtrc(chi_val,len(f_obs)-1),f_obs
```

```python
# a,b,f_obs=chi_square_test(bins,y)
# b
```

With 100 deviates and intervals of range(-3,3,1) the chi-square value is around 87 and p-value, is>0.05, (for 5% significance test). Thus, the deviates belong to gaussian. Even with 100 deviates but intervals of range(-3,3,0.1) the chi-squared value on an average increases.However, the p value >0.05, thus deviates are from normal distribution.

with 500 deviates and interval (-3,3,1), the chisquare value is 471, but it also show p>0.05, thus these are gaussian and with interval (-3,3,0.1) there is also an increase in the chi square value and p>0.05 so samples belong from gaussain.

With finer intervals chi-square value increases and even with higher no. of samples the chi-square increases.

```python
def new_generate_normal_deviate(i):   
    y = []
    for _ in range(int(0.999*i)):
        rsq,v1,v2 =0,0,0    
        while (rsq>=1 or rsq ==0):
            x1,x2 = np.random.uniform(0,1,2)
            v1 = 2*x1 -1
            v2 = 2*x2 -1
            rsq = v1**2 + v2**2
        fac = np.sqrt(-2*np.log(rsq)/rsq)
        y.append(v2*fac)
    
    y.append(v1*fac*500)
    return y
```

```python
no_sample=900
y = new_generate_normal_deviate(no_sample)
```

```python
_,p_chi = chisquare(y)
_,p_KS=ks_1samp(y,stats.norm.cdf)
critical_val = 1.358*np.sqrt(2*len(y)/(len(y)**2))
```

```python
print(f"The p value from chisquare is {p_chi} for {no_sample} samples\nand p from KS test is {p_KS}, with critical value {critical_val}")
```
