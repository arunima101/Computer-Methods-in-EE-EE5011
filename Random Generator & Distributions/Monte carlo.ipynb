```python
import numpy as np
from math import *
import scipy.special as sp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import random
```

```python
def u(x,y):
    alpha=np.pi*np.sin(10*(np.sqrt(x**2+y**2)-0.5))

    uval=x*np.cos(alpha)+y*np.sin(alpha)-0.5
    return uval
```

```python
def v(x,y):
    alpha=np.pi*np.sin(10*(np.sqrt(x**2+y**2)-0.5))

    vval=-x*np.sin(alpha)+y*np.cos(alpha)-0.5
    return vval
```

```python
def new_func(x,y):
    return u(x,y)**2+v(x,y)**2
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

X, Y = np.meshgrid(x, y)
Z = new_func(X, Y)
```

```python
def normalize(x):
    return (x-x.min())/(x.max()-x.min())
```

```python
samples_in_vol =0
samples_out_of_volume = 0
for i in range(100000):
    x1,x2 = 4*np.random.rand(2)-2   
    val = new_func(x1,x2)
    if(abs(val)<1):
        samples_in_vol+=1
    else:
        samples_out_of_volume+=1
```
